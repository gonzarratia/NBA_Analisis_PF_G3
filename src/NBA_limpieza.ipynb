# Limpieza de tabla players_info

```python
df_common_player_info = pd.read_csv("C:/Users/Celia V/Desktop/HENRY/PROYECTO FINAL/common_player_info.csv")
df_common_player_info.head()
```

```python
df_common_player_info.info()
```

```python
df_player = pd.read_csv("C:/Users/Celia V/Desktop/HENRY/PROYECTO FINAL/player.csv")
df_player.head()
```

```python
df_player.info()
```

```python
df_merged = df_player.merge(
    df_common_player_info,
    left_on="id",
    right_on="person_id",
    how="left"
)
df_merged.info()
```

```python
df_final = df_merged[[
    "id",              # este será tu player_id
    "first_name_x",    # viene de df_player
    "last_name_x",     # viene de df_player
    "full_name",       # viene de df_player
    "is_active",       # viene de df_player
    "birthdate",       # viene de df_common_player_info
    "country",         # viene de df_common_player_info
    "height",          # viene de df_common_player_info
    "weight"           # viene de df_common_player_info
]]

# Renombramos columnas 
df_final = df_final.rename(columns={
    "id": "player_id",
    "first_name_x": "first_name",
    "last_name_x": "last_name"
})

print(df_final.head())
```

```python
df_final.info()
```

```python
df_final["birthdate"] = pd.to_datetime(df_final["birthdate"], errors="coerce")

# Verificar
print(df_final[["player_id", "birthdate"]].head())
print(df_final["birthdate"].dtype)
```

```python
# Contar cuántos NaT hay en la columna birthdate
nat_count = df_final["birthdate"].isna().sum()

print("Cantidad de NaT en birthdate:", nat_count)
```

```python
nulos = df_final.isna().sum()
print(nulos)
```

```python
# Rellenar nulos en first_name con la primera palabra de full_name
df_final["first_name"] = df_final.apply(
    lambda row: row["full_name"].split()[0] if pd.isna(row["first_name"]) else row["first_name"],
    axis=1
)

# Verificamos nulos
print(df_final["first_name"].isna().sum())
```

```python
# Reemplazar nulos en country por "Other"
df_final["country"] = df_final["country"].fillna("Other")

# Verificar nulos
print(df_final["country"].isna().sum())
```

```python
# Función para convertir "6-10" en centímetros
def height_to_cm(h):
    try:
        feet, inches = h.split("-")
        feet = int(feet)
        inches = int(inches)
        total_inches = feet * 12 + inches
        return total_inches * 2.54  # convertir a cm
    except:
        return None  # si hay nulos o valores inválidos

# Aplicar la función a la columna height
df_final["height"] = df_final["height"].apply(height_to_cm)

# Verificar
print(df_final[["player_id", "height"]].head())
print(df_final["height"].dtype)  # debería ser float64
```

```python
# Convertir de cm a metros y redondear a 2 decimales
df_final["height"] = (df_final["height"] / 100).round(2)

# Verificar
print(df_final[["player_id", "height"]].head())
print(df_final["height"].dtype)  # debería ser float64
```

```python
# Convertir de libras a kilogramos 
df_final["weight"] = (df_final["weight"] * 0.453592).round(2)

# Verificar
print(df_final[["player_id", "weight"]].head())
print(df_final["weight"].dtype)  # debería ser float64
```

```python
# Calcular los promedios ignorando nulos
mean_height = df_final["height"].mean()
mean_weight = df_final["weight"].mean()

# Reemplazar nulos con el promedio
df_final["height"] = df_final["height"].fillna(mean_height)
df_final["weight"] = df_final["weight"].fillna(mean_weight)

# Verificar que ya no haya nulos
print("Nulos en height:", df_final["height"].isna().sum())
print("Nulos en weight:", df_final["weight"].isna().sum())
```

```python
df_final.to_csv("C:/Users/Celia V/Desktop/HENRY/PROYECTO FINAL/players_info.csv", index=False)
```
